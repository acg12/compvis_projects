```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
```

Image can be found here: [source](https://raw.githubusercontent.com/acg12/compvis_projects/main/Point-Based%20Processing/img.jpg)

```python
image = cv2.imread('img.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
print(image.shape)
```

```python
# Visualize original histogram

def plot_histogram(image, title):
  intensity = np.zeros(256, dtype=int)
  height = image.shape[0]
  width = image.shape[1]

  for i in range(height):
    for j in range(width):
      intensity[image[i][j]] += 1

  plt.plot(intensity)
  plt.title(title + ' Histogram')
  plt.xlabel('Intensity')
  plt.ylabel('Level')
  plt.show()

plot_histogram(image, 'Original Image')
```

```python
cv2_imshow(image)
```

# Gray-Level Slicing

### First approach (left picture)

Kita akan meng-highlight grey-level yang tinggi (convert jadi putih/255) di sebuah bagian tertentu dalam image (range [A, B]) dan menurunkan semua grey-level yang lainnya menjadi level yang konstan (convert jadi hitam/0).

image.png

```python
modified_1 = image.copy()
height = modified_1.shape[0]
width = modified_1.shape[1]

for i in range(height):
  for j in range(width):
    pixel = modified_1[i][j]

    if pixel >= 120 and pixel <= 220:
      modified_1[i][j] = 255
    else:
      modified_1[i][j] = 0

plot_histogram(image, 'Original Image')
plot_histogram(modified_1, 'Modified Image')
```

```python
cv2_imshow(modified_1)
```

### Second approach

Kita akan meng-highlight grey-level yang tinggi (convert jadi putih/255) di sebuah bagian tertentu dalam image (range [A, B]) dan membiarkan grey-level di bagian image yang lain tetap seperti semula (preserve other levels).

```python
modified_2 = image.copy()
height = modified_2.shape[0]
width = modified_2.shape[1]

for i in range(height):
  for j in range(width):
    pixel = modified_2[i][j]

    if pixel >= 120 and pixel <= 220:
      modified_2[i][j] = 255

plot_histogram(image, 'Original Image')
plot_histogram(modified_2, 'Modified Image')
```

```python
cv2_imshow(modified_2)
```
